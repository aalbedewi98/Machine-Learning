```python
# ============================================
# LINEAR REGRESSION EXAMPLE: PREDICTING CO2 EMISSIONS
# ============================================

# Step 0: Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# -------------------------------
# Step 1: Create the dataset
# -------------------------------
# We simulate a small dataset with car Weight (kg), Engine Volume (cm3), and CO2 emission (g/km)
data = {
    'Weight': [2300, 2500, 1800, 2000, 2100],
    'Volume': [1300, 1400, 1000, 1200, 1250],
    'CO2': [250, 270, 200, 220, 230]
}

# Convert the dictionary into a pandas DataFrame
df = pd.DataFrame(data)

# Display the dataset
print("Dataset preview:")
print(df)

# -------------------------------
# Step 2: Prepare input and output
# -------------------------------
# X: features (independent variables)
X = df[['Weight', 'Volume']]

# y: target (dependent variable)
y = df['CO2']

# -------------------------------
# Step 3: Create and train the model
# -------------------------------
# Initialize the Linear Regression model
regr = LinearRegression()

# Train the model using the dataset
regr.fit(X, y)

# -------------------------------
# Step 4: Make predictions
# -------------------------------
# Predict CO2 emission for a new car using a DataFrame (to avoid warnings)
new_car = pd.DataFrame([[2300, 1300]], columns=['Weight', 'Volume'])
predicted_CO2 = regr.predict(new_car)

print("\nPredicted CO2 emission for a car with weight 2300kg and volume 1300cm3:")
print(predicted_CO2[0])

# -------------------------------
# Step 5: Analyze the model
# -------------------------------
# Coefficients of the model (impact of each feature)
print("\nModel coefficients:")
print("Weight coefficient:", regr.coef_[0])
print("Volume coefficient:", regr.coef_[1])

# Intercept of the model
print("Intercept:", regr.intercept_)

# -------------------------------
# Step 6: Visualize the data (optional)
# -------------------------------
# Since this is 2D features, we can plot CO2 vs Weight and CO2 vs Volume separately
plt.figure(figsize=(12,5))

# CO2 vs Weight
plt.subplot(1,2,1)
plt.scatter(df['Weight'], y, color='blue', label='Actual CO2')
plt.xlabel('Weight (kg)')
plt.ylabel('CO2 Emission (g/km)')
plt.title('CO2 vs Weight')
plt.legend()

# CO2 vs Volume
plt.subplot(1,2,2)
plt.scatter(df['Volume'], y, color='green', label='Actual CO2')
plt.xlabel('Engine Volume (cm3)')
plt.ylabel('CO2 Emission (g/km)')
plt.title('CO2 vs Volume')
plt.legend()

plt.tight_layout()
plt.show()

# ============================================
# Explanation:
# 1. Created a small dataset with car features (Weight, Volume) and target (CO2).
# 2. X contains the features and y contains the target variable.
# 3. LinearRegression() fits a plane to predict CO2 based on Weight and Volume.
# 4. regr.predict() allows predicting CO2 for new input values.
# 5. Using a DataFrame with column names avoids UserWarning.
# 6. regr.coef_ shows how strongly each feature affects CO2.
# 7. Visualization helps understand relationships between each feature and CO2.
# ============================================
```

# Coefficient

The coefficient is a factor that describes the relationship with an unknown variable.
In this case, we can ask for the coefficient value of weight against CO2, and for volume against CO2. The answer(s) we get tells us what would happen if we increase, or decrease, one of the independent values.

```python
print(regr.coef_)
```

The result array represents the coefficient values of weight and volume.

Weight: 0.00755095
Volume: 0.00780526

These values tell us that if the weight increase by 1kg, the CO2 emission increases by 0.00755095g.

And if the engine size (Volume) increases by 1 cm3, the CO2 emission increases by 0.00780526 g.

I think that is a fair guess, but let test it!

We have already predicted that if a car with a 1300cm3 engine weighs 2300kg, the CO2 emission will be approximately 107g.

What if we increase the weight with 1000kg (from 2300 to 3300) what will be the CO2 emission?

Ans: 107.2087328 + (1000 * 0.00755095) = 114.75968

```python
predictedCO2 = regr.predict([[3300, 1300]])

print(predictedCO2)
```
