```python
# Unit 9: Convolutional Neural Network (CNN) for Object Recognition using CIFAR-10
# Author: Your Name
# This code implements a CNN, trains it, and evaluates its performance on the CIFAR-10 dataset.

# ===================== IMPORTS AND SEEDS =====================
import numpy as np
import tensorflow as tf
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import pandas as pd
from keras.preprocessing.image import array_to_img
from sklearn.metrics import classification_report, confusion_matrix
from IPython.display import display

# Set random seeds for reproducibility
np.random.seed(888)
tf.random.set_seed(112)

# ===================== LOAD DATASET =====================
# CIFAR-10 dataset: 60,000 32x32 RGB images across 10 classes
(x_train_all, y_train_all), (x_test, y_test) = cifar10.load_data()

# Class labels
LABEL_NAMES = ['airplane', 'automobile','bird','cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# ===================== DATA EXPLORATION =====================
print("Training data shape:", x_train_all.shape)
# (50000, 32, 32, 3) → 50k images, 32x32 pixels, 3 color channels (RGB)

# Display first image using IPython
display(array_to_img(x_train_all[0]))

# Display using matplotlib
plt.imshow(x_train_all[0])
plt.show()

# Check label of first image
print("Label index:", y_train_all[0][0])
print("Label name:", LABEL_NAMES[y_train_all[0][0]])

# ===================== PREPROCESS DATA =====================
# Scale pixel values to [0,1] for faster training
x_train_all = x_train_all / 255.0
x_test = x_test / 255.0

# Convert labels to categorical (one-hot encoding)
y_cat_train_all = to_categorical(y_train_all, 10)
y_cat_test = to_categorical(y_test, 10)

# ===================== CREATE VALIDATION SET =====================
VALIDATION_SIZE = 10000  # 20% for validation
x_val = x_train_all[:VALIDATION_SIZE]
y_val_cat = y_cat_train_all[:VALIDATION_SIZE]

# Remaining 40,000 images for training
x_train = x_train_all[VALIDATION_SIZE:]
y_cat_train = y_cat_train_all[VALIDATION_SIZE:]

print("Training set shape:", x_train.shape)
print("Validation set shape:", x_val.shape)

# ===================== BUILD CNN MODEL =====================
model = Sequential()

# --- FIRST CONVOLUTION + POOLING LAYER ---
model.add(Conv2D(filters=32, kernel_size=(4,4), input_shape=(32,32,3), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

# --- SECOND CONVOLUTION + POOLING LAYER ---
model.add(Conv2D(filters=32, kernel_size=(4,4), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

# --- FLATTEN AND DENSE LAYERS ---
model.add(Flatten())  # Flatten 3D output to 1D
model.add(Dense(256, activation='relu'))  # Hidden dense layer
model.add(Dense(10, activation='softmax'))  # Output layer: 10 classes

# Compile model with categorical crossentropy loss and Adam optimizer
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()  # Shows model architecture and parameters

# ===================== TRAIN MODEL WITH EARLY STOPPING =====================
early_stop = EarlyStopping(monitor='val_loss', patience=2)  # Stop if val_loss doesn't improve for 2 epochs

# Fit the model
history = model.fit(x_train, y_cat_train, epochs=25, validation_data=(x_val, y_val_cat), callbacks=[early_stop])

# Convert training history to DataFrame for visualization
metrics = pd.DataFrame(history.history)

# Plot Loss and Accuracy
metrics[['loss', 'val_loss']].plot()
plt.title('Training Loss vs Validation Loss', fontsize=16)
plt.show()

metrics[['accuracy', 'val_accuracy']].plot()
plt.title('Training Accuracy vs Validation Accuracy', fontsize=16)
plt.show()

# ===================== EVALUATE ON TEST DATA =====================
test_loss, test_acc = model.evaluate(x_test, y_cat_test)
print(f"Test Loss: {test_loss}, Test Accuracy: {test_acc}")

# ===================== CLASSIFICATION REPORT AND CONFUSION MATRIX =====================
# Predict classes
predictions = np.argmax(model.predict(x_test), axis=-1)

# Print classification report
print(classification_report(y_test, predictions, target_names=LABEL_NAMES))

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, predictions)
print(conf_matrix)

# ===================== PREDICT SINGLE IMAGE =====================
# Display a single test image
plt.imshow(x_test[16])
plt.show()

my_image = x_test[16].reshape(1,32,32,3)  # Reshape to match input shape
predicted_class = np.argmax(model.predict(my_image), axis=-1)

print("Predicted class index:", predicted_class[0])
print("Predicted class name:", LABEL_NAMES[y_test[16][0]])
```
