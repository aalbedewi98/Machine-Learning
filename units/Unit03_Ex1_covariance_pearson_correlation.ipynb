```python
# --- Importing necessary libraries ---
from numpy import mean, std, cov
from numpy.random import randn, seed
from matplotlib import pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns

# --- Step 1: Set random seed for reproducibility ---
seed(1)  # ensures the same random numbers are generated each run

# --- Step 2: Generate random datasets ---
# 'data1' is a normally distributed variable with mean around 100 and std deviation around 20
data1 = 20 * randn(1000) + 100

# 'data2' is related to data1 but with added noise (+50 shift + random variation)
data2 = data1 + (10 * randn(1000) + 50)

#  TRY CHANGING this relationship to see the effect:
# Example 1:
#data2 = data1 * 2 + (10 * randn(1000))    # Stronger positive correlation

# Example 2:
#data2 = -data1 + (10 * randn(1000))       # Negative correlation

# Example 3:
#data2 = 10 * randn(1000) + 50             # Weak/no correlation

# --- Step 3: Calculate covariance matrix ---
covariance = cov(data1, data2)  # shows how both variables vary together

# --- Step 4: Calculate Pearson correlation coefficient ---
corr, _ = pearsonr(data1, data2)  # ranges from -1 (negative) to +1 (positive)

# --- Step 5: Visualize relationship ---
plt.figure(figsize=(6,4))
plt.scatter(data1, data2, alpha=0.5)
plt.title(f'Scatter Plot (Correlation = {corr:.2f})')
plt.xlabel('data1')
plt.ylabel('data2')
plt.show()

# --- Step 6: Summarize numerical results ---
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearson correlation: %.3f' % corr)
```
